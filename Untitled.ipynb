# Initial Data Generation

```python
t_init = 0
t_final = 50
dt = 0.001
t = torch.arange(start=t_init, end=t_final, step=dt)
omega = 0.3
phi = 3.143/3



pos_x = torch.sin(omega*t+phi)
pos_y = torch.cos(omega*t+phi)
pos = torch.t(torch.stack((pos_x, pos_y, torch.zeros_like(pos_x))))

vel_x = omega*torch.cos(omega*t+phi)
vel_y = -omega*torch.sin(omega*t+phi)
vel = torch.t(torch.stack((vel_x, vel_y, torch.zeros_like(vel_x))))

ang_vel = torch.tensor([[0,0,omega]]).repeat(t.size()[0], 1)

accel_x = -omega*omega*torch.sin(omega*t+phi)
accel_y = -omega*omega*torch.sin(omega*t+phi)
accel = torch.t(torch.stack((accel_x, accel_y, -torch.ones_like(accel_x)*9.81)))
```
